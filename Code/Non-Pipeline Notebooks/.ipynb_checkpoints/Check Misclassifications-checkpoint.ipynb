```python
import numpy as np
import csv
import textwrap
from dateutil.parser import parse
from operator import add

import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir) 
```

## Find misclassified RIDs for PET

```python
#find PET misclassified patients
results_PET = np.loadtxt('results_PET.csv', delimiter=",")
indices = np.loadtxt(parentdir + '/test_indices.csv', delimiter=",")

miss_indices = []
for line in range(len(results_PET)):
    if results_PET[line][0] != results_PET[line][1]:
        miss_indices.append([indices[line],results_PET[line][0],results_PET[line][1]])
        
RID_list = np.loadtxt(parentdir + '/LongitudinalDataAnalysis.csv', delimiter=",", skiprows = 1)

miss_RID = []
for i in miss_indices:
    miss_RID.append([int(RID_list[int(i[0])][0]), i[1],i[2]])

just_RIDS_PET = sorted([i[0] for i in miss_RID])
print "# Misclassified is", len(miss_RID)
print "Misclassified patients have RID numbers:", just_RIDS_PET
```

# Find Misclassified RIDs 

```python
results = np.loadtxt(parentdir + '/results.csv', delimiter=",")
indices = np.loadtxt(parentdir + '/test_indices.csv', delimiter=",")

miss_indices = []
for line in range(len(results)):
    if results[line][0] != results[line][1]:
        miss_indices.append([indices[line],results[line][0],results[line][1]])
        
RID_list = np.loadtxt(parentdir + '/LongitudinalDataAnalysis.csv', delimiter=",", skiprows = 1)

miss_RID = []
for i in miss_indices:
    index_of_incorrect = int(i[0])
    patient_RID = int(RID_list[index_of_incorrect][0])
    miss_RID.append([patient_RID, i[1],i[2]])
    
just_RIDS = [i[0] for i in miss_RID]
print "# Misclassified is", len(miss_RID)
print "Misclassified patients have RID numbers:", just_RIDS
```

```python
list(set(just_RIDS_PET).intersection(just_RIDS))
```

## Look at "Truth" Information in ADNIMERGE/DXSUM

```python
patient_dict = {}
labels = []

RID = 0
COLPROT = 4
EXAMDATE = 6
DX_bl = 7
DX = 51

with open('../../Data___Database/ADNIMERGE.csv') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        if row[0] == "RID":
            labels = row
        else:
            if int(row[0]) in just_RIDS:
                if int(row[0]) not in patient_dict:
                    patient_dict[int(row[0])] = [[row[COLPROT], row[EXAMDATE], row[DX_bl], row[DX]]]
                else:
                    patient_dict[int(row[0])].append([row[COLPROT], row[EXAMDATE], row[DX_bl], row[DX]])
            
patient_diagnosis_dict = {}
with open('../../Assessments/DXSUM_PDXCONV_ADNIALL.csv') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader) #skip header
    for row in reader:
        RID = int(row[2])
        EXAMDATE = row[8]
        DXOTHDEM = row[47]
        DX_curren = row[10]
        Dx_change = row[9]
        if int(RID) in just_RIDS:
            if RID not in patient_diagnosis_dict:
                patient_diagnosis_dict[RID] = []
            patient_diagnosis_dict[RID].append([EXAMDATE, DXOTHDEM, DX_curren, Dx_change])
            
                    
DX_CURREN = {'1':'NL', "2": 'MCI', "3": 'AD', "":""}
DX_CHANGE = {'1':"Stable:NL to NL",'2':"Stable: MCI to MCI",'3':"Stable: AD to AD",'4':"Conv:NL to MCI",
            '5':"Conv:MCI to AD",'6':"Conv:NL to AD", '7':"Rev:MCI to NL",'8':"Rev:AD to MCI",'9':"Conv:AD to NL","":""}
    
```

## Print Out Details For Each Patient 

```python
for i in patient_dict:
    print "Patient is: ", i
    prefix =  "      "
    preferredWidth = 70
    
    wrapper = textwrap.TextWrapper(initial_indent=prefix, width=preferredWidth,
                               subsequent_indent=" "*len(prefix))
    
    predicted = miss_RID[just_RIDS.index(i)][1]
    actual = miss_RID[just_RIDS.index(i)][2]
    
    studies = [j[0] for j in patient_dict[i]]
    exam_dates = [j[1] for j in patient_dict[i]]
    DX_bl = patient_dict[i][0][2]
    progression = [j[3] for j in patient_dict[i]]
    
    
    ##Look in Diagnosis Dict
    diag_dates = [j[0] for j in patient_diagnosis_dict[i]]
    dxothdem = [j[1] for j in patient_diagnosis_dict[i]]
    dx_curren = [DX_CURREN[j[2]] for j in patient_diagnosis_dict[i]]
    dx_change = [DX_CHANGE[j[3]] for j in patient_diagnosis_dict[i]]
    diag = map(add, dx_curren, dx_change)
    
    message =  "-Predicted was: "+ str(predicted)+ " Actual was: "+ str(actual)
    print wrapper.fill(message)
    message =  "-Baseline DX: "+ str(DX_bl)
    print wrapper.fill(message)
    message =  "-Studies: "+ str(studies)
    print wrapper.fill(message)
    message =  "-Exam Dates: "+ str(exam_dates)
    print wrapper.fill(message)
    message =  "-Progression of Disease: "+ str(progression)
    print wrapper.fill(message)
    print " "
    message = "-Diagnosis file dates: "+ str(diag_dates)
    print wrapper.fill(message)
    message = "-Diag: "+ str(diag)
    print wrapper.fill(message)
    print " "

```
