```python
results = pd.read_csv('ConsistencyTrials.csv')

averages = {}
std = {}
for imp in ['meanmode','knn','softimpute']:
    for feat in ['none', 'AffinityPropagation', 'SVD']:
        curr_result = results[(results['Imputation Method'] == imp) & (results['Feature Selection'] == feat)]
        averages[imp+"_"+feat] = list(curr_result.mean())
        std[imp+"_"+feat] = list(curr_result.std())
```

```python
def compare_models(averages, metric):
    results = []
    for a in averages:
        curr = averages[a]
        
        if metric == 'accuracy':
            result = [curr[4*i] for i in range(len(curr)/4)]
        if metric == 'recall':
            result = [curr[4*i+1] for i in range(len(curr)/4)]
        if metric == 'precision':
            result = [curr[4*i+2] for i in range(len(curr)/4)]
        if metric == 'f1':
            result = [curr[4*i+3] for i in range(len(curr)/4)]
        
        result.insert(0,a)
        results.append(result)
    
    with open(str(metric)+'_Results.csv','wb') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow([' ','Logistic','RandomForest','kNN','MLP','SVM','AdaBoost','GradientBoosting','LinearSVM','DecisionTree'])
        for value in results:
            writer.writerow(value)
    return results
        
```

```python
def consider_standard_deviations(std, metric):
    results = []
    for a in std:
        curr = std[a]
        
        if metric == 'accuracy':
            result = [curr[4*i] for i in range(len(curr)/4)]
        if metric == 'recall':
            result = [curr[4*i+1] for i in range(len(curr)/4)]
        if metric == 'precision':
            result = [curr[4*i+2] for i in range(len(curr)/4)]
        if metric == 'f1':
            result = [curr[4*i+3] for i in range(len(curr)/4)]
        
        result.insert(0,a)
        results.append(result)
    
    with open(str(metric)+'_Results_StandardDeviations.csv','wb') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow([' ','Logistic','RandomForest','kNN','MLP','SVM','AdaBoost','GradientBoosting','LinearSVM','DecisionTree'])
        for value in results:
            writer.writerow(value)
            
    return results
        
```

```python
def plot_model_difference(metric):
    results = pd.read_csv('ConsistencyTrials.csv')

    averages = {}
    std = {}
    for imp in ['meanmode','knn','softimpute']:
        for feat in ['none', 'AffinityPropagation', 'SVD']:
            curr_result = results[(results['Imputation Method'] == imp) & (results['Feature Selection'] == feat)]
            averages[imp+"_"+feat] = list(curr_result.mean())
            std[imp+"_"+feat] = list(curr_result.std())
    
    averages = compare_models(averages, metric)
    stds = consider_standard_deviations(std, metric)
    
    order = [' ','LoReg','RF','kNN','MLP','SVM','ABoost','GBoost','LinSVM','Tree']    
    average_unraveled = []
    std_unraveled = []
    
    for i in averages:
        parameter_type = i[0]
        for j in range(len(i)):
            if j != 0:
                model_type = order[j]
                average_unraveled.append([parameter_type+"__"+model_type, i[j]])
                
    for i in stds:
        parameter_type = i[0]
        for j in range(len(i)):
            if j != 0:
                model_type = order[j]
                std_unraveled.append([parameter_type+"__"+model_type, i[j]])
    

    imp_types = set([i[0][:i[0].find("_")] for i in average_unraveled])
    for t in imp_types:
        y = [i[1] for i in average_unraveled if i[0][:i[0].find("_")] == t]
        x = np.array(range(1,len(y)+1))
        x_names = [i[0][i[0].find("__")+2:] for i in average_unraveled if i[0][:i[0].find("_")] == t]
        e = [i[1] for i in std_unraveled if i[0][:i[0].find("_")] == t]
        c = {'none': 'green', 'SVD': 'blue', 'AffinityPropagation':'red'}
        col = [c[i[0][i[0].find("_")+1:i[0].find("__")]] for i in average_unraveled if i[0][:i[0].find("_")] == t]
        
        plt.scatter(x, y, c=col, s=50, zorder=3)
        plt.errorbar(x, y, e, linestyle='None')
        plt.title(str(t)+' '+str(metric))
        plt.axis([0, 30, 0.60, 1])
        plt.xticks(x, x_names, size='small')
        plt.xticks(rotation=70, ha = 'right')

        classes = ['none','SVD','Affinity Propagation']
        class_colours = ['green','blue','red']
        recs = []
        for i in range(0,len(class_colours)):
            recs.append(mpatches.Rectangle((0,0),0.5,0.5,fc=class_colours[i]))
        plt.legend(recs,classes,bbox_to_anchor=(1.52, 0.8))

        #plt.savefig(str(t)+' '+str(metric)+'.png')
        plt.show()
        
    
    feature_types = set([i[0][i[0].find("_")+1:i[0].find("__")] for i in average_unraveled])
    
    for t in feature_types:
        y = [i[1] for i in average_unraveled if i[0][i[0].find("_")+1:i[0].find("__")] == t]
        x = np.array(range(1,len(y)+1))
        x_names = [i[0][i[0].find("__")+2:] for i in average_unraveled if i[0][i[0].find("_")+1:i[0].find("__")] == t]
        e = [i[1] for i in std_unraveled if i[0][i[0].find("_")+1:i[0].find("__")] == t]
        c = {'LoReg':'red','RF':'green','kNN':'blue','MLP':'black','SVM':'yellow','ABoost':'magenta','GBoost':'cyan','LinSVM':'orange','Tree':'white'}
        col = [c[i[0][i[0].find("__")+2:]] for i in average_unraveled if i[0][i[0].find("_")+1:i[0].find("__")] == t]
        
        markers = {'meanmode':'^','knn':'o','softimpute':'s'}
        cluster = [markers[i[0][:i[0].find("_")]] for i in average_unraveled]
        fig, ax = plt.subplots()

        for xp, yp, m, co in zip(x, y, cluster, col):
            ax.scatter([xp],[yp], marker=m, c = co, s = 50, zorder = 3)

        ax.errorbar(x, y, e, linestyle='None')
        plt.title(str(t)+' '+str(metric))
        plt.axis([0, 30, 0.60, 1])
        plt.xticks(x, x_names, size='small')
        plt.xticks(rotation=70, ha = 'right')

        classes = [a for a in c]
        class_colours = [c[a] for a in c]
        recs = []
        for i in range(0,len(class_colours)):
            recs.append(mpatches.Rectangle((0,0),0.5,0.5,fc=class_colours[i]))
        plt.legend(recs,classes,bbox_to_anchor=(1.3, 1))

        
        #plt.savefig(str(t)+' '+str(metric)+'.png')
        plt.show()
     
```

```python
plot_model_difference('recall')
```

```python
order = [' ','Logistic Regression','Random Forest','kNN','MLP','SVM','AdaBoost','GradientBoost','LinearSVM','Decision Tree']  
averages_and_stds = []
for i in range(len(averages)):
    for j in range(len(averages[i])):
        if j == 0:
            header = averages[i][j]
        else:
            averages_and_stds.append([header, order[j], averages[i][j], stds[i][j]])
            
averages_and_stds.sort(key=lambda x: x[2], reverse = True)
print averages_and_stds
for modeltype1 in range(len(averages_and_stds)):
    better_than = []
    for modeltype2 in range(i,len(averages_and_stds)):
        if averages_and_stds[modeltype2][2]+averages_and_stds[modeltype2][3] < averages_and_stds[modeltype1][2]-averages_and_stds[modeltype1][3]:
            better_than.append(averages_and_stds[modeltype2][0] + averages_and_stds[modeltype2][1])
```

```python
### Average across feature type

no_feature_selection = [i[2] for i in averages_and_stds if i[0][i[0].find("_")+1:] == 'none']
SVD_feature_selection = [i[2] for i in averages_and_stds if i[0][i[0].find("_")+1:] == 'SVD']
AP_feature_selection = [i[2] for i in averages_and_stds if i[0][i[0].find("_")+1:] == 'AffinityPropagation']

print "No feature selection",sum(no_feature_selection)/len(no_feature_selection)
print "SVD feature selection",sum(SVD_feature_selection)/len(SVD_feature_selection)
print "AP feature selection",sum(AP_feature_selection)/len(AP_feature_selection)
```
